# Draw a square

```
initializeTurtle(initial_speed=6, initial_window_size=(600,210))
penup(); setx(100); sety(150); pendown()
color("yellow")
for side in [1, 2, 3, 4]:
    forward(100)
    right(90)
    
penup(); setx(300); sety(150); pendown()
color("red")
forward(100)
right(90)
color("orange")
forward(100)
right(90)
color("yellow")
forward(100)
right(90)
color("green")
forward(100)
```

# Draw a pentagon

```
initializeTurtle(initial_speed=6, initial_window_size=(600,210))
    
penup(); setx(100); sety(150); pendown()
color("blue")
i = 5
angle = 360 / i
for _ in range(i):
    forward(100)
    right(angle)
    
penup(); setx(350); sety(150); pendown()
color("yellow")
i = 8
angle = 180 - 360 / 8
for _ in range(i):
    forward(100)
    right(angle)
```

```
initializeTurtle(initial_speed=8, initial_window_size=(600,210))
penup(); setx(100); sety(100); pendown()
for length in [10, 20, 30, 40, 50, 60]:
    forward(length)
    right(90)

penup(); setx(300); sety(100); pendown()
color('purple')
for angle in [10, 20, 30, 40, 50, 60]:
    forward(20)
    right(angle)

penup(); setx(500); sety(180); pendown()    
color("red")
width(5)
angles = [-90, 0, 0, -90,
          135, 0, 0, 0, 
          90, 0, 0, 0,
          135, -90, 0, 0,
          90, 0, 0, 0]
for angle in angles:
    # Turn right, then go forward 25.
    # (How far to turn?
    #  Use the angle variable!)
    right(angle)
    forward(25)    
```

```
initializeTurtle(initial_speed=8, initial_window_size=(600,150))
width(5)

# Move back without drawing anything
right(90)
penup()
forward(-140)
pendown()

# Draw three shapes of different colors, with space in between
for prettycolor in ["red", "orange", "yellow"]:
    color(prettycolor)
    for _ in range(4):
        forward(50)
        right(90)
    penup()
    forward(100)
    pendown()
```

# Draw a rainbow

```
import math
initializeTurtle(initial_speed=10, initial_window_size=(600,250))
width(10); right(90)

rainbow = ["red", "orange", "yellow", "green", "blue", "purple"]

# Write whatever code you want here!
for i, c in enumerate(rainbow): 
    color(c)
    x0 = 51
    y0 = 51 + 10 * i
    penup(); setx(x0); sety(y0); pendown()
    for i in range(51, 51+200):
        x = i
        y = round(math.sqrt(100**2 - (x-150)**2)) + y0
        setx(x); sety(y)
```

# Draw a spirangle

```
initializeTurtle(initial_speed=10, initial_window_size=(600,250))
width(1); right(90); color("blue")
for side in range(20):
    forward(side*10)
    right(120)
```

# Draw a spiral

```
def spiral(n=100):
    initializeTurtle(initial_speed=10, 
                     initial_window_size=(400,400))
    width(3); right(90); color("green")    
    for i in range(n):
        forward(i)
        right(20)
    
spiral()
```
